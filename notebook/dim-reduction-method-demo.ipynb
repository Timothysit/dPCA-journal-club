```python
trial_dim = 0
neuron_dim = 1
stimuli_dim = 2
time_dim = 3
num_trials = num_trial_per_stimuli * num_stimuli

pop_firing_rate = np.zeros(shape=(num_trials, num_neurons, num_stimuli, num_time_points))

trial_counter = 0
for stimuli in np.arange(num_stimuli):
    for trial in np.arange(num_trial_per_stimuli):
        for cluster_1_neuron in np.arange(cluster_1_num_neuron):
            for time in np.arange(num_time_points):
                random_noise = np.random.normal(loc=0, scale=1)
                cluster_1_firing_rate = cluster_1_time_latent[time] + \
                                        cluster_1_stimuli_latent[stimuli] + \
                                        random_noise
                pop_firing_rate[trial_counter, cluster_1_neuron, stimuli, time] = cluster_1_firing_rate
    trial_counter = trial_counter + 1
    
    
```
